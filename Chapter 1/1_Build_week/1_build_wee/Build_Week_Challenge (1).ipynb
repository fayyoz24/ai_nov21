```python
# Import all the required libraries

import requests
import pandas as pd
import numpy as np
import matplotlib as plt
from bs4 import BeautifulSoup
```

```python
page = requests.get("https://www.goodreads.com/list/show/1.Best_Books_Ever")
print(page)
```

```python
soup = BeautifulSoup(page.content, "html.parser")
```

```python
book_url = soup.find_all('a', {'itemprop': 'url'})
book_url
```

```python
url_list = []
book_url_list = []
for link in soup.find_all('a', {'class': 'bookTitle'}):
    url_list.append((link.get('href')))
for x in url_list:
    book_url_list.append('https://goodreads.com' + x)
book_url_list
```

```python
# Scraping the Author

authors = soup.find_all('a', class_='authorName')

author_names = []

for author in authors:
    author_names.append(author.get_text())

author_names
```

```python
# Scraping the Book Title

titles = soup.find_all('a', class_='bookTitle')

title_list = []
book_titles = []

for title in titles:
    title_list.append(title.get_text())
    
for x in title_list:
    book_titles.append(((x.split('\n'))[1].split('\n')[0]))

book_titles
```

```python
for url in book_url_list:
    print(url)
```
