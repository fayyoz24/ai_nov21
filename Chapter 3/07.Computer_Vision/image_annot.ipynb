```python
import cv2
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('./pngs/dog_cat.jpg')
img0 = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
```

```python
cv2.rectangle(img0, (20, 30), (250, 400), (0, 255, 0), 2)
cv2.rectangle(img0, (263, 15), (465, 400), (0, 255, 0), 2)
cv2.rectangle(img0, (480, 25), (720, 400), (0, 255, 0), 2)

cv2.putText(img0, 'dog!', (20, 30), cv2.FONT_ITALIC, 1, (255,0,0), 1)
cv2.putText(img0, 'cat!', (263, 15), cv2.FONT_ITALIC, 1, (255,0,0), 1)
cv2.putText(img0, 'dog!', (480, 25), cv2.FONT_ITALIC, 1, (255,0,0), 1)
plt.figure(figsize=(20,15))
plt.imshow(img0)
```
